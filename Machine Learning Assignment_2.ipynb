```r
lm.fit <- lm(ZN ~ CRIM, data = data_31)

# Plot the data points
plot(data_31$CRIM, data_31$ZN, 
     main = "Linear Regression of ZN vs CRIM", 
     xlab = "CRIM - crim per capita crime rate by town.", 
     ylab = "ZN -  proportion of residential land zoned for lots over 25,000 sq.ft.", 
     pch = 19,  # solid circle
     col = "black")  # color of points

# Add the regression line to the plot
abline(lm.fit, col = "orange", lwd = 2)  # lwd = line width
```

```r
# Fit the linear model
lm.fit <- lm(ZN ~ CRIM, data = data_31)

# Generate the 4-panel diagnostic plot
par(mfrow = c(2, 2))  # Set up the plotting area to be 2x2 panels
plot(lm.fit)
```

```r
plot(hatvalues(lm.fit))
```

```r
lm.fit <- lm(ZN ~ RM + AGE + MEDV, data = data_31)
summary(lm.fit)
```

```r
predicted_values <- predict(lm.fit)
actual_values <- data_31$ZN

# Scatter plot of Actual vs Predicted
plot(actual_values, predicted_values, 
     main = "Actual vs Predicted Values", 
     xlab = "Actual ZN", 
     ylab = "Predicted ZN", 
     pch = 19, 
     col = "red")

# Add a 45-degree line for reference
abline(a = 0, b = 1, col = "black", lwd = 2) 
```
